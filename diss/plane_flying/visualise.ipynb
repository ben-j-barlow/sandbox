```python
import numpy as np
import matplotlib.pyplot as plt

DEG_TO_RAD = np.pi / 180

# Function to update positions and heading
def update_position(x, y, psi, phi, v, dt):
    psi_dot = g * np.tan(phi) / v
    x_dot = v * np.cos(psi)
    y_dot = v * np.sin(psi)
    psi += psi_dot * dt
    x += x_dot * dt
    y += y_dot * dt
    return x, y, psi


# Simulation loop
def run_and_plot_simulation(total_time, dt, v=20, g=9.81):
    # Initial conditions
    x, y, psi = 0, 0, 0  # initial positions and heading angle
    phi = 0  # initial bank angle

    # Lists to store trajectory
    x_list, y_list = [x], [y]

    for _ in range(int(total_time / dt)):
        phi_change = np.random.choice([-5, 5]) * DEG_TO_RAD
        phi += phi_change
    
        x, y, psi = update_position(x, y, psi, phi, v, dt)
    
        x_list.append(x)
        y_list.append(y)

    # Plotting the trajectory
    plt.figure(figsize=(10, 6))
    plt.plot(x_list, y_list, label="Trajectory")
    plt.xlabel("X Position (meters)")
    plt.ylabel("Y Position (meters)")
    plt.title("Aircraft Trajectory")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
# dt =  0.1

run_and_plot_simulation(total_time=100, dt=0.1, v=20, g=9.81)
```

```python
# dt = 0.5 
# much smoother curves

run_and_plot_simulation(total_time=100, dt=0.5, v=20, g=9.81)
```
